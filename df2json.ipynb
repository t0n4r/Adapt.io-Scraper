```python
import json
import pandas as pd
```

```python
company_df = pd.read_csv('company_df.csv')
company_index_df = pd.read_csv('company_index_df.csv') 
company_contacts_df = pd.read_csv('company_contacts_df.csv') 
similar_company_df = pd.read_csv('similar_company_df.csv')
```

```python
# Good to go

company_contacts = company_contacts_df.groupby(['company_name']).apply(lambda x: x[['contact_name', 'contact_jobtitle', 'contact_email_domain', 'contact_department']].to_dict('records')).reset_index().rename(columns={0:'contact_details'})
```

```python
company_contacts[company_contacts['company_name'].str.contains("Avtec")]
```

```python
company_contacts
```

```python
similar_company_df
```

```python
similar_company = similar_company_df.groupby(['company_name']).apply(lambda x: x[['similar_company']].to_dict('records')).reset_index().rename(columns={0:'similar_companies'})
```

```python
similar_company
```

```python
dff = pd.merge(company_df, company_contacts, on="company_name")
```

```python
dff = pd.merge(dff, similar_company, on='company_name')
```

```python
dff
```

```python
# Create json files

dff = dff.to_dict('records')

json_object = json.dumps(dff, indent = 4)

print(json_object)
with open('company_profiles.json', 'w') as f:
    f.write(json_object)
```

```python
# Create Json Files

company_index = company_index_df.to_dict('records')

json_object = json.dumps(company_index, indent = 4)

print(json_object)
with open('company_index.json', 'w') as f:
    f.write(json_object)
```
